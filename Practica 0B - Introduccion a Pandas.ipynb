### Slicing

Hay veces que queremos particionar o filtrar el dataset. Para ello, Pandas proporciona distintas funciones para hacerlo de manera eficiente

```python
# Vamos a coger un muestreo aleatorio como hemos visto anteriormente para trabajar en esta seccion
sample_df = df.sample(10, random_state=seed)
sample_df
```

#### `loc`

La función `loc` escoge los elementos de acuerdo con su índice. Esto quiere decir que, si es un muestreo aleatorio del Dataframe, necesitamos conocer los valores de los índices. 

```python
# Si queremos coger el valor de la primera fila del dataset conocemos que su indice es 6252
sample_df.loc[6252]
```

#### `iloc`

Por otro lado, la función `iloc` escoge los elementos de acuerdo con su indice relativo. Esto quiere decir que para escoger un elemento no necesitamos saber su indice sino su posición

```python
# Si queremos coger el valor de la primera fila del dataset conocemos que su posicion es 0
sample_df.iloc[1]
```

Estos métodos funcionan igual que el _slicing_ en listas en Python, es decir, permite coger rangos, definir saltos, etc...

```python
# Para coger las 5 primeras filas del DataFrame
sample_df.iloc[:5]
```

```python
# O sabiendo que la columna de teoria es la segunda (indice 1)
sample_df.iloc[:,1]
```

### Condiciones

Además de indexar el Dataframe, una de las funciones más potentes de Pandas es que permite filtrar los Dataframes por condiciones utilizando la función `loc`

```python
# Si queremos sacar todos los alumnos con teoria aprobada
df.loc[(df['Teoria'] > 5)]
```

Para añadir múltiples condiciones, tienen que ir englobadas en paréntesis e ir seguidas de un `&` o un `|`, ya que son operadores lógicos, no booleanos

```python
# Si queremos sacar todos los alumnos con teoria y practica aprobadas
df.loc[(df['Teoria'] > 5) & (df['Practica'] > 5)]
```

### Eficiencia

Una de las mayores razones por la que usar una librería u otra en cada momento es por eficiencia. Numpy es mucho más eficiente que Pandas a la hora de iterar sus datos, algo que se realizará en múltiples ocasiones en esta asignatura. 

```python
from time import perf_counter
df_np = df.to_numpy()
```

```python
# Ver que nota de teoria es mayor que practicas -> Pandas iterrows()
counter = 0

perf = perf_counter()
for i,row in df.iterrows():
    if row['Teoria'] > row['Practica']:
        counter += 1
perfPd = perf_counter() - perf

print(counter, "notas. ", "Pandas iterrows():", perfPd.__round__(5), "segundos")
```

```python
# Ver que nota de teoria es mayor que practicas -> Numpy bucle for
counter = 0

perf = perf_counter()
for _, t, p, _,  _ in df_np:
    if t > p:
        counter += 1
perfNp = perf_counter() - perf

print(str(counter) + " notas.  Numpy: " + str(perfNp.__round__(5)) + " segundos")
```

Sin embargo, más importante aún que elegir usar Pandas o Numpy es el uso de operaciones vectorizadas siempre que sea posible. Python es un lenguaje MUY lento, principalmente al no ser compilado, por lo que cada bucle `for` que nos podamos ahorrar reemplazándolo por alguna operación vectorizada programada internamente en `C` hará que nuestro código vaya mucho más rápido.

```python
# Ver que nota de teoria es mayor que practicas -> Pandas vectorizado
counter = 0

perf = perf_counter()
counter = df[df['Teoria'] > df['Practica']].shape[0]
perfPdV = perf_counter() - perf

print(counter, "notas. ", "Pandas:", perfPdV.__round__(5), "segundos")
```

```python
# Ver que nota de teoria es mayor que practicas -> Numpy vectorizado
counter = 0

perf = perf_counter()
counter = np.sum(df_np[:,1] > df_np[:,2])
perfNpV = perf_counter() - perf

print(str(counter) + " notas.  Numpy vectorizado: " + str(perfNpV.__round__(5)) + " segundos")
```

```python
# Algunas métricas
print("Ganancia de velocidad de Numpy sobre Pandas (bucle for):", (perfPd/perfNp).__round__(2))
print("Ganancia de velocidad de Numpy sobre Pandas (vectorizado):", (perfPdV/perfNpV).__round__(2))

print("\nGanancia de velocidad de Pandas al vectorizar:", (perfPd/perfPdV).__round__(2))
print("Ganancia de velocidad de Numpy al vectorizar:", (perfNp/perfNpV).__round__(2))
```
